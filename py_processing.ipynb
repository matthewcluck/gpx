```python
def process_hike(hike):
    coordinates = []
    
    for track in hike.tracks:
        for segment in track.segments:
            for point in segment.points:
                coordinates.append((point.longitude, point.latitude))

    hike_df = pd.DataFrame(coordinates, columns=['longitude', 'latitude'])
    hike_gdf = gpd.GeoDataFrame(hike_df, geometry=gpd.points_from_xy(hike_df.longitude, hike_df.latitude), crs="EPSG:4326")
                
    return hike_gdf
            
                
processed_hike = process_hike(first_hike_gpx)
m = processed_hike.explore(tiles="OpenStreetMap")
m
```
